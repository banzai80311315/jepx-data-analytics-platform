# 環境構築ログ（Day1）

本日は JEPX 電力価格分析プロジェクトの **開発環境構築**を実施した。  
Linux VM 上で Python データ分析環境とプロジェクト構造を整備した。

# 1. 開発環境

今回構築した開発環境は以下の通り。
```
Windows (ホスト)
│
├ Git
├ VS Code
└ VirtualBox
↓
Ubuntu VM (開発環境)
│
├ Python 3.10
├ venv (仮想環境)
├ pandas
├ matplotlib
└ JEPX分析プロジェクト
```
# 2. Git リポジトリ

GitHub 上にプロジェクトを作成し、VM に clone。
```bash
git clone <repository>
```

VM 上の作業ディレクトリ
```bash
~/workspace/jepx-data-analytics-platform
```

# 3. プロジェクト構造

以下のディレクトリ構造を作成。

```
jepx-data-analytics-platform
├ app
│ ├ loader.py
│ ├ processor.py
│ ├ analyzer.py
│ └ visualizer.py
│
├ data
│ ├ raw
│ │ └ spot_summary_2023.csv
│ └ processed
│
├ notebooks
├ outputs
│ └ figures
│
├ logs
├ test
│ └ test_pipeline.py
│
├ requirements.txt
├ .gitignore
└ README.md
```

# 4. 仮想環境

Python 仮想環境を作成。
```bash
python3 -m venv venv
```
起動
```bash
source venv/bin/activate
```

# 5. 依存ライブラリ管理

依存ライブラリを `requirements.txt` で管理。

```
pandas
numpy
matplotlib
seaborn
scikit-learn
statsmodels
plotly
jupyter
streamlit
```

インストール
```bash
pip install -r requirements.txt
```

# 6. .gitignore

Git管理対象外のファイルを設定。
```
Python

pycache/
*.pyc

virtual environment

venv/

Jupyter

.ipynb_checkpoints/

logs

logs/

processed data

data/processed/*
```
# 7. テストコード

CSV 読み込み確認のためのテストスクリプトを作成。

```
test/test_pipeline.py
```
# 8. テスト実行

VM 上で実行。
```bash
python test/test_pipeline.py
```
