# gitのブランチ戦略(Day3)

## gitのブランチ戦略
- main
- feature
- fix

### main
実行環境がpullするための安定ブランチ

### feature
新規機能開発ブランチ

### fix
不具合修正用ブランチ

### ルール
- 実行環境はmainしか使わない
- 作業は必ずfeature/fix
- mainに入れるのは「反映してもよいものだけ」

**push**
```bash
git status
git add .
git commit -m "banzai"
git push
```

**merge**
```bash
git switch main
git pull origin main
git merge feature/csv-import
git push origin main
```

**delete**

ローカル
```bash
git branch -d feature/csv-import
```

リモート
```bash
git push origin --delete feature/csv-import
```

## gitの基本
pullしたら何が起きるかの確認
```bash
git fetch
git log main..origin/main
git diff main origin/main
git diff --name-only main origin/main
git pull
```

### logとは
```bash
git log --oneline --graph
```

### headとは
現在作業している場所を示すポインタ
```bash
git branch
```

### fetchとは
リモートの更新を取得する
ただし自分のブランチは変更しない（リモート追跡が更新され、ローカルは更新されない）

```bash
git fetch
```

すると、リモート追跡とローカルでの差分が確認ができる
```bash
git diff
```

**定番操作**
リモートとローカルの差分を見る
```bash
git diff main origin/main
```

リモートが何コミット進んでるか
```bash
git log main..origin/main
```

ローカルが何コミット進んでいるか
```bash
git log origin/main..main
```

変更ファイルの確認
```bash
git diff --name-only main origin/main
```

### upstreamとは
すでに存在するブランチ(ローカル)へ切替
```bash
git switch feature/csv-import
```

新しくブランチ(ローカル)を作成して切替
```bash
git switch -c feature/csv-import
```

set upstream for git pull/status

```text
ローカル
feature/csv-import

追跡先
origin/feature/csv-import
```
この関係をupstream（上流ブランチ）と呼ぶ。

upstreamが存在すると
```bash
git pull origin feature/csv-import
↓ 省略
git pull
```

upstreaの確認
```bash
git branch -vv
```

### ブランチ削除
ローカル
```bash
git branch -d feature/csv-import
```

リモート
```bash
git push origin --delete feature/csv-import
```
